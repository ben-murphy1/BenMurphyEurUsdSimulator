```python
# The program will ask the user to input a starting price for their EURUSD.
# It will use exception handling in a 'while loop' to handle any incorrect inputs from the user.
# Occassionally there are time.sleep(x) delays to make the output more realistic and manageable.


# Importing packages
import random
random.seed()
import time

while True:
    try:
        eurusd_price = float(input("Please enter an FX Price for EURUSD e.g. 1.0680: "))
        break  # Exit the loop if the input is valid
    except ValueError:
        print("Invalid input! Please enter a positive integer.")

print(f"\nThank you, your EURUSD FX price of '{eurusd_price}' has been accepted.")
time.sleep(1)

starting_price = eurusd_price
tick = random.random()


# The FX prices will print out, one by one as they 'trade' until they rise or fall by 15 points.
# The program uses a double if/else while loop inside a function to create a list of outputs which have 'traded'.


print(f"\nWelcome to The Foreign Exchange Market simulator!\nThe Price will move up and down until it rises or falls 15 points (+- 0.0015)")
time.sleep(1)
print("Your trading will end at either {:.4f} or {:.4f}".format(eurusd_price + 0.0015, eurusd_price - 0.0015))
time.sleep(1)
print("WARNING: Although highly unlikely it is conceivable that the trading could continue\nfor an extended period due to mathematical variance and the random nature of the price moves.")
time.sleep(1)
print("\nWE ARE NOW TRADING!")

def create_list():
    '''This function will randomly create the stock prices we are going to use going forward'''
    eurusd_price_list = []
    while True:
        tick = random.random()
        if tick > 0.5:
            global eurusd_price
            eurusd_price += 0.0001
            eurusd_price_list.append(eurusd_price)
            print(f"Higher! {round(eurusd_price, 4)}")  # Round to 4 decimal places
            time.sleep(0.25)
            if eurusd_price > (starting_price + 0.0014):
                print("EURUSD has stopped trading for the day because it has already gone up 15 points!")
                break

        else:
            eurusd_price -= 0.0001
            eurusd_price_list.append(eurusd_price)
            print(f"Lower! {round(eurusd_price, 4)}")  # Round to 4 decimal places
            time.sleep(0.25)
            if eurusd_price < (starting_price - 0.0014):
                print("EURUSD has stopped trading for the day because it has already fallen 15 points!")
                break

    return eurusd_price_list

prices = create_list()
```

```python
# Next we will format the Traded Prices list into nice clean 4 decimal places.
traded_prices = [ round(elem, 4) for elem in prices ]
print(f"Your list of rounded traded prices in order of appearance is:\n{traded_prices}")

# Number of traded Prices
length = len(traded_prices)
print(f"\nWow! In total there were {length} seperate pric changes.")
```

```python
# Since charts are a critical part of trading we want to see the data visually.
# Here we will graph the movement of the stock price through out the day.

import matplotlib.pyplot as plt

xs = [x for x in range(len(traded_prices))]

# The X axis is the number of seperate prices.
# The Y axis is the price

plt.plot(xs, traded_prices)
plt.xlabel('# Of Trades')
plt.ylabel('Stock Price')
plt.title('Traded Prices')
plt.show()
plt.close()

# We also are interested in seeing the frequency with which each individual price traded.

plt.hist(traded_prices, bins = 30) # 30 bins as there are 30 conceivable prices that could trade.
plt.title('Histogram of Prices Traded')
plt.xlabel('Stock Price')
plt.ylabel('Number of Trades')
plt.show()
```

```python
# The following section of code uses different techniques such as functions and math's to analyse the traded prices list.
# There is a lot of data in this section, we will sort it afterwards.

# Highest Price of the Day
highest_price = max(traded_prices)
print(f"The highest share price of the day is: {highest_price}")

# Lowest Price of the Day
lowest_price = min(traded_prices)
print(f"\nThe lowest share price of the day is: {lowest_price}")


# First Traded Price
opening_price = traded_prices[0]
print(f"\nThe first traded price is: {opening_price}")


# Last Traded Price
closing_price = traded_prices[-1]
print(f"\nThe last traded price is: {closing_price}")


# Most Common Price & Frequency
from statistics import mode
most_common_price = mode(traded_prices)
count = traded_prices.count(most_common_price)

print(f"\nThe most common individual share price of the day was {most_common_price} which traded {count} times.")


# Average Traded Price, rounded.
from statistics import mean
average_price = mean(traded_prices)
ave = round(average_price,2) # These secondary variables will be used in the next section
print(f"\nThe average share price of the day was {round(average_price,4)}.")


# Finding Above Average Prices
def mean(traded_prices):
    if len(traded_prices) == 0:
        return 0.0
    return sum(traded_prices) / len(traded_prices)

def grt_mean(traded_prices):
    m = mean(traded_prices)
    return [i for i in traded_prices if i > m]

print(f"\nThe traded prices above the average, {round(average_price,2)}, are :\n{grt_mean(traded_prices)}")


# Calculating Percentage of time above average
percentage_above = (len(grt_mean(traded_prices)) / len(traded_prices)) * 100
pa = round(percentage_above,2)
print(f"\nIn total, {round(percentage_above,2)}% of all traded prices were above the average.")

# Calculating Percentage of time below average
def grt_mean(traded_prices):
    m = mean(traded_prices)
    return [i for i in traded_prices if i < m]

print(f"\nThe traded prices below the average, {round(average_price,2)}, are :\n{grt_mean(traded_prices)}")

percentage_below = (len(grt_mean(traded_prices)) / len(traded_prices)) * 100
ba = round(percentage_below,2)
print(f"\nIn total, {round(percentage_below,2)}% of all traded prices were below the average.")

# Calculating Percentage of time above starting price
above_starting = [i for i in traded_prices if i > starting_price]
print(f"\nThe prices traded above the start price were:\n{above_starting}")

a = (len(above_starting) / len(traded_prices)) * 100
print(f"\nThe stock traded above the starting price {round(a,2)}% of the time")
asr = round(a,2)

# Calculating Percentage of time below starting price
below_starting = [i for i in traded_prices if i < starting_price]
print(f"\nThe prices traded below the starting price were:\n{below_starting}")

b = (len(below_starting) / len(traded_prices)) * 100
print(f"\nThe stock traded below the starting price {round(b,2)}% of the time")
bsr = round(b,2)


print(f"\nThe remaining % was when the price traded at the start price.")


# Checks if the stock rose or fell by the end of the day using an if/else function.
def stock_price_advance(a,b):
    """Checks if a < b"""
    if a < b:
        print("\nThe share price advanced on the day as the closing price was greater than the opening price.")
    else :
        print("\nThe share price failed to advance on the day as the closing price was less than the opening price.")

stock_price_advance(opening_price,closing_price)
```

```python
# Now we have lots of data on the prices which the stock has traded.
# Up until now we have had long lists of data.
# To make it digestible we will create some dictionaries of the findings to store it in a functional way.

traded_prices_dict = {
    "High of the day": highest_price,
    "Low of the day": lowest_price,
    "Starting Price": starting_price,
    "Most Common Price": most_common_price,
    "Occurances": count,
    "Average Price": ave,
    "% Above Average": pa,
    "% Below Average": ba,
    "% Above Start": asr,
    "% Below Start": bsr,
}

print("Trading Statistics:")
# Prints the dictionary vertically for readability
for key, value in traded_prices_dict.items():
    print(key, ' : ', value)
```

```python
# Elements we do not have in our dictionary are the longest consecutive rise or fall as well as the average consecutive rise and fall.
# Here we will find these rises and falls and then add them to our dictionary.

def longest_rising_streak(traded_prices): # Finding the Longest Consecutive Rise
    longest_count = 0
    current_count = 1

    for i in range(1, len(traded_prices)):
        if traded_prices[i] >= traded_prices[i-1]:
            current_count += 1
        else:
            longest_count = max(longest_count, current_count)
            current_count = 1

    return max(longest_count, current_count)

rising_streak = longest_rising_streak(traded_prices)
print(f"Longest rising streak is {rising_streak}.")

def longest_falling_streak(traded_prices): # Finding the longest consecutive fall
    start = 0
    falling_count = 1

    for i in range(1, len(traded_prices)):
        if traded_prices[i] <= traded_prices[i-1]:
            falling_count += 1
        else:
            start = max(start, falling_count)
            falling_count = 1

    return max(start, falling_count)

falling_streak = longest_falling_streak(traded_prices)
print(f"Longest falling streak is {falling_streak}.")

def average_length_of_rising_streaks(traded_prices): # Finding the average rise
    count = 0
    total_count = 0
    previous_value = traded_prices[0]  # Assuming the list has at least one element

    for current_value in traded_prices[1:]:
        if current_value >= previous_value:
            count += 1
        else:
            total_count += count
            count = 0
        previous_value = current_value

    if count > 0:
        total_count += count

    num_of_streaks = traded_prices.count(traded_prices[0])  # Assuming the list starts with a rising streak
    average_length_rise = total_count / num_of_streaks

    return average_length_rise

average_length_rise = average_length_of_rising_streaks(traded_prices)
print(f"The average length of rising streaks is: {average_length_rise:0.2f}")


def average_length_of_falling_streaks(traded_prices): # Finding the average Fall
    count = 0
    total_count = 0
    previous_value = traded_prices[0]  # Assuming the list has at least one element

    for current_value in traded_prices[1:]:
        if current_value <= previous_value:
            count += 1
        else:
            total_count += count
            count = 0
        previous_value = current_value

    if count > 0:
        total_count += count

    num_of_streaks = traded_prices.count(traded_prices[0])  # Assuming the list starts with a falling streak
    average_length_fall = total_count / num_of_streaks

    return average_length_fall

average_length_fall = average_length_of_falling_streaks(traded_prices)
print(f"The average length of falling streaks is: {average_length_fall:0.2f}")


# Now we will add this to our dictionary
traded_prices_dict["Longest Rise"] = rising_streak
traded_prices_dict["Longest Fall"] = falling_streak
traded_prices_dict["Average Rise"] = average_length_rise
traded_prices_dict["Average Fall"] = average_length_fall

print("\nThe New Dictionary is:\n")
# and re-print the dictionary to show the amendment
for key, value in traded_prices_dict.items():
    print(key, ' : ', value)
```

```python
# If we wanted a more general chart instead of the granular detail above.
# We could slice the 'traded_prices' list to give us every fourth data point.
# This would still be highlighting the overall direction of the stock.

sliced_trades = traded_prices[1:-1:4]
print(f"The sliced down price list is:\n{sliced_trades}")

# First we will reprint the original chart from above.
# Then we will recreate the chart with the smaller list 'sliced_trades'
# They will give the same overall direction.
print("\nHere are the two charts side by side:")
xs = [x for x in range(len(traded_prices))]

plt.plot(xs, traded_prices)
plt.xlabel('# Of Trades')
plt.ylabel('Stock Price')
plt.title('Detailed Traded Prices')
plt.show()
plt.close()

xs = [x for x in range(len(sliced_trades))]

plt.plot(xs, sliced_trades)
plt.xlabel('# Of Trades')
plt.ylabel('Stock Price')
plt.title('Sliced Traded Prices')
plt.show()
plt.close()


```
